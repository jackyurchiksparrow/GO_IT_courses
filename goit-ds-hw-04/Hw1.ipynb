10. Знайдіть повторювані записи (від 2-го випадку повторення і далі) у заданому масиві і позначте їх як True. Перше повторення має бути False.

Input np.random.seed(100) 

a = np.random.randint(0, 5, 10)

[0 0 3 0 2 4 2 2 2 2] 

Output [False True False True False False True True True True]

```python
np.random.seed(100)                 # noqa: NPY002
a = np.random.randint(0, 5, 10)     # noqa: NPY002

print(f"a = {a}")
vals, first_indices = np.unique(a, return_index=True) # first indices where elements occur
result = np.full(len(a), True)
result[first_indices] = False
result
```
